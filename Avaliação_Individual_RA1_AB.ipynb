# Questão 1: 
Um elemento radioativo tem uma meia-vida de 5 anos. Se inicialmente existem 500 gramas desse elemento, faça um gráfico que mostre como a quantidade desse elemento varia com o tempo e determine quantos gramas restarão após 15 anos. Utilize a equação $A = A0 * e^{-kt}$, onde $A0$ é a quantidade inicial do elemento, $k$ é a constante de decaimento (em 1/anos) e $t$ é o tempo decorrido em anos.



```python
import matplotlib.pyplot as plt
import numpy as np

A0 = 500
k = np.log(2)/5

t = np.arange(0, 16)


A = A0 * np.exp(-k*t)
print('Restarão' ,A[-1], 'gramas')
plt.plot(t, A)

plt.grid(True)

plt.xlabel('Anos')
plt.ylabel('Quantidade')

plt.show()
```

# Questão 2
Usando, a seu critério as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x)=2^x + 3^x + 25$

2. $f(x)=\log_2(x) - \log_2(3) - 1$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)

f = np.log2(x) - np.log2(3) -1

y = f

plt.plot(x, y)

plt.grid(True)

plt.title('Função Crescente')
plt.xlabel('x')
plt.ylabel('f(x)=log2(x) - log2(3) - 1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2**x + 3**x +25

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y)

plt.grid(True)

plt.title('Função Crescente')
plt.xlabel('x')
plt.ylabel('$f(x) = 2^x + 3^x +25$')

plt.show()
```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 2} \frac{x^3 - 8}{x - 2}$

2. $\lim_{x \to \infty} \frac{3x^2 + 2x - 1}{x^2 - 4x + 7}$



$$ \lim_{x \to \infty} \frac{3x^2 + 2x -1}{x^2 - 4x + 7} = \lim_{x \to \infty} \frac{3 + \frac{2}{x} - \frac{1}{x^2}}{1 - \frac{4}{x} + \frac{7}{x^2}} = \frac{3}{1} = 3 $$



```python
import numpy as np
import sympy
import matplotlib.pyplot as plt

def grafico_limite():
    def f(x):
        return (3*x**2 + 2*x -1)/(x**2 - 4*x + 7)

    x = np.linspace(-10, 10, 1000)
    y = f(x)

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('f(x)')

    limite = sympy.limit(f(x), x, sympy.oo)
    print(f"Limite: {limite}")

    plt.grid( True)
    plt.show()

grafico_limite()
```

$$ \lim_{x \to 2} \frac{x^3 - 8}{x - 2} = \lim_{x \to 2} \frac{(x-2)(x^2 + 2x + 4)}{x-2} = \lim_{x \to 2} (x^2 + 2x + 4) = 12 $$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy

def f(x):
    return (x**3 - 8)/(x-2)


x = sympy.symbols('x')


limite = sympy.limit(f(x), x, 2)

print(f"Limite: {limite}")


x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.grid(True)
plt.show()
```
